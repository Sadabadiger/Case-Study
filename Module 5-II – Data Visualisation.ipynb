```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("BigMartSalesData.csv")
df1=df.copy()
df1=df1.drop(df[df.Year!=2011].index)
df1=df1.groupby('Month').agg({'Amount':np.sum})
df1.plot(kind='line',color='red')
```

2.Plot Total Sales Per Month for Year 2011 as Bar Chart. Is Bar Chart Better to visualize than Simple Plot?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("BigMartSalesData.csv")
df1=df.copy()
df1=df1.drop(df[df.Year!=2011].index)
df1=df1.groupby('Month').agg({'Amount':np.sum})
df1.plot(kind='bar',color='Green')
```

**After comparing these two charts, bar charts are good to analyse rather than line graph**

3. Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("BigMartSalesData.csv")
df1=df.copy()
df1=df1.drop(df[df.Year!=2011].index)
df1=df1.groupby('Country').agg({'Amount':np.sum})
plt.figure(figsize=(10,10))
plt.pie(df1.values,labels=df1.index,autopct='%1.1f%%')
plt.show()
```

4. Plot Scatter Plot for the invoice amounts and see the concentration of amount. In which range most of the invoice amounts are concentrated

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("BigMartSalesData.csv")
df1=df.copy()
df1=df1.drop(df[df.Year!=2011].index)
df1=df1.groupby('InvoiceNo').agg({'Amount':np.sum})
df1
plt.scatter(df1.values,df1.values,color='magenta')
plt.grid(True)
plt.show()
```
